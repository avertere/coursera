```r
#The basic R installation makes quite a few data sets available to us
help(co2)
co2
class(co2) #time series object
plot(co2, main='Atmospheric CO2 Concentration')
```

```r
# We  can  calculate the  slope  and  interceptvalues  
# “by  hand”  as  follows.  First,  to  get  the 
# concentrations and times they were collected 
# available to compute with, we can type
co2.values = as.numeric(co2)
co2.times = as.numeric(time(co2))
SSxx  = sum((co2.times-mean(co2.times))*(co2.times -mean(co2.times)))
SSxy  = sum((co2.values-mean(co2.values))*(co2.times -mean(co2.times)))
(slope = SSxy/SSxx)
(intercept= mean(co2.values)-slope*mean(co2.times))
```

```r
#lm im a more simply way
co2.linear.model = lm(co2 ~ time(co2))
plot(co2, main='Atmospheric CO2 Concentrationwith Fitted Line')
abline(co2.linear.model)
```

```r
# If you’d like to compute your residuals by hand, you can calculate the fitted values and subtract them from your observed responses with a command like
co2.fitted.values = slope*co2.times + intercept
co2.residuals = co2.values -co2.fitted.values
# You might be a little lazier and would like R to do this for you. We “interrogate” the linear model with a command like
(co2.residuals = resid( co2.linear.model))
```

```r
# checkin lm supositions: normality for erros and homocedasticity
par(mfrow=c(1,3))
(c02.residuals = resid(co2.linear.model))
hist(co2.residuals, main="Histogram of CO2 Residuals")
qqnorm(c02.residuals, main= "Normal Probability Plot") 
qqline(c02.residuals)
plot(c02.residuals~time(co2), main="Residuals on Time")
```

```r
#Thistleton and SadigovReviewing Basic StatisticsWeek 1Practical Time Series AnalysisPage 6It  seems  silly  not  to  also  zoom  in  on  the  residuals  using  the  “xlim”  argument  to  see  theseasonality in the data set. We will address the oscillations in the data later in the course.
plot(c02.residuals ~ time(co2), xlim=c(1960, 1963), main="Zoomed in Residuals on Time")
```

```r
# inferential statistics
help(sleep)
plot(extra~group, data=sleep, main = "Extra Sleep in Gossett Data by Group")
attach(sleep)
extra.1=extra[group==1]
extra.2=extra[group==2]
t.test(extra.1,extra.2,paired=TRUE, alternative="two.sided")
```

*Lembrete:* The statistic we have used is the Student-t for the paired differences $𝑥_d=x-y$ 

$$t=\frac{\bar{x}_d-0}{s_d/\sqrt{n}}$$

The  rather  large  negative  t-value  leads  us  to  believe  that  the  visual  evidence  in  the  box  plot  is supported  by  the  statistical  approach.  The  $p$-value  is  quite  small  and  our  data  are  significant  at the customary $\alpha=0.01$ level.  

$$𝑝<\alpha~ \Rightarrow~ \text{reject}$$

The  sample  average  difference  obtained  from  these  ten  individuals  estimates  the  population parameter $\mu_d$. An interval estimate provides a quality statement and is more interesting than the point  estimate  of   $\bar{x}−\bar{y}=−1.58$. To that, we use

$$\bar{d}\pm t_{\alpha/2}\frac{s}{\sqrt{n}}$$

At  the  $95\%$  level  of  confidencewe  make  the  traditional statement that, whatever the actual difference in means, our interval estimate is 

$$𝐶𝐼(\mu_d, 0.95):\quad [−2.4598858;0.7001142]$$

Of course, for the test to be meaningful, we assume that the population of differences is normally distributed. 

```r
diffs= extra.1-extra.2
qqnorm(diffs, main= "Normal Probability Plot")
qqline(diffs)
```

**weak stationary:**
- not systematic change in mean (no trend)
- not systematic change in variance
- no periodic variations

**autocovariance function**

$$\gamma(s,t)=Cov(X_s,X_t)=E[(X_s-\mu_s)(X_t-\mu_t)]; \quad 
\gamma(t,t)=E[(X_t-\mu_t)^2]=\sigma^2_t$$

Under stationary time series,  no matter $t$, we will consider autocovariance function 

$$\gamma_k=\gamma(t,t+k)\sim c_k$$

where $c_k$ is a estimation of $\gamma_k$ given by

$$c_k=\frac{\sum_{t=1}^{N-k}(x_t-\bar{x})(x_{t+k}-\bar{x})}{N}$$

with $\bar{x}=\frac{\sum_{t=1}^Nx_t}{N}$.

```r
# in R we can acf() routine to have autocovariance estimaions
purely_random_process = ts(rnorm(1000))
print(purely_random_process)
(acf(purely_random_process, type='covariance'))
```

**autocorrelation function**

Again, under weak stationarity,

$$\rho_k:=\rho(X_t,X_{t+k}):=\frac{\gamma_k}{\gamma_0}\in[-1,1]$$

which will be estimated by

$$r_k=\frac{c_k}{c_0}=\frac{\sum_{t=1}^{N-k}(x_t-\bar{x})(x_{t+k}-\bar{x})}{\sum_{t=1}^N(x_t-\bar{x})^2}.$$

With them we can produze the correlogram using again the acf routine.

```r
purely_random_process = ts(rnorm(1000))
(acf(purely_random_process, main='correlogram'))
```

random wal model with white noise as residual

$$X_t = X_{t-1}+Z_t,\quad  Z_t i.i.d N(\mu,\sigma^2)$$

It is easy to see that $X_t=\sum_{i=1}^tZ_i$, wich lead to $E(z_t)= \mu t$ and $Var(Z_t)=t\sigma^2$ 

```r
# rw with no stationarity
x = cumsum(rnorm(1000))
rw = ts(x)
plot(rw, main='A random walk')
acf(rw)
```

But $\nabla X_t:=X_t-X_{t-1}=Z_t$, the difference operator, generates a stationary time series.

```r
# remove the trend in rw using the difference operatorb
plot(diff(rw))
acf(diff(rw))
```

**Introduction to Moving Average Processes - $MA(q)$:**

$$X_t=Z_t+\theta_1Z_{t-1}+\ldots +\theta_qZ_{t-q},$$

with $Z_i's$ i.i.d with $Z_1\sim N(\mu,\sigma^2)$.

```r
# simulate MA(2) given by X_t=Z_t+0.7Z_{t-1}+0.2Z_{t-2}

# noise
noise = rnorm(10000)

# ma model
ma_2=NULL
for (i in 3:10000){
    ma_2[i]=noise[i]+0.7*noise[i-1]+0.2*noise[i-2]
}

#shift data to left by 2 units
moving_avergade_process=ma_2[3:10000]

# ts structure
moving_avergade_process=ts(moving_avergade_process)

par(mfrow=c(2,1))
plot(moving_avergade_process, main="A moving average process of order 2")
acf(moving_avergade_process, main = "Correlogram of a moving average process of order 2")
```
